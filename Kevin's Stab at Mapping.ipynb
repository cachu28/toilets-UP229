```python
import pandas as pd
import geopandas as gpd
import contextily as ctx
import matplotlib.pyplot as plt
```

```python
toilets = gpd.read_file('toilets.geojson').to_crs(epsg='3857')
```

```python
toilets.head()
```

```python
counts = gpd.read_file('homelesscount_tracts.geojson').to_crs(epsg='4326')
```

```python
counts.head()
```

```python
counts = counts[['homeless_count', 'geometry']]
```

```python
counts.head()
```

```python
#LA county wide map

f, ax = plt.subplots(figsize=(20,20))

toilets.plot('type', ax=ax, marker='o', markersize=20, legend=True)

counts.plot('homeless_count', 
            ax=ax, 
            cmap='plasma', 
            legend=True, 
            legend_kwds={'orientation': 'vertical'}, 
            alpha=0.3, 
            label='Count')
ax.set_xlim([-118.721783, -118.08]) 
ax.set_ylim([33.68, 34.4])
ax.set_title('2020 Homeless Count Data with Santitation Stations in LA County', fontsize=25)

ctx.add_basemap(ax,crs='EPSG:4326', source=ctx.providers.Stamen.TonerLite)

plt.show()
```

```python
# Central LA specific map

f, ax = plt.subplots(figsize=(20,20))

toilets.plot('type', ax=ax, marker='o', markersize=20, legend=True)

counts.plot('homeless_count', 
            ax=ax, 
            cmap='plasma', 
            legend=True, 
            legend_kwds={'orientation': 'vertical'}, 
            alpha=0.25, 
            label='Count')
ax.set_xlim([-118.40, -118.20]) 
ax.set_ylim([34.003307, 34.121767])
ax.set_title('2020 Homeless Count Data with Santitation Stations in Central LA', fontsize=25)

ctx.add_basemap(ax,crs='EPSG:4326', source=ctx.providers.Stamen.TonerLite)

plt.show()

```
