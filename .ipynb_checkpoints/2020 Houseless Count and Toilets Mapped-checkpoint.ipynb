## 2020 Houseless Count and Toilets Mapped

```python
import pandas as pd
import geopandas as gpd
import contextily as ctx
import matplotlib.pyplot as plt
```

```python
toilets = gpd.read_file('MiscData/toilets.geojson').to_crs(epsg=3857)
```

```python
toilets.sample(10)
```

```python
counts = gpd.read_file('MiscData/2020countbytract.geojson')
```

```python
counts.head()
```

```python
# Plotting both toilets and counts on the same map

f, ax = plt.subplots(figsize=(20,10))

counts.plot(ax=ax, cmap='plasma', legend=True, legend_kwds={'orientation': 'horizontal'}, alpha=0.3)

toilets.plot(ax=ax, marker='o', color='red', markersize=20)

ax.set_xlim([-118.721783, -118.08]) 
ax.set_ylim([33.68, 34.4])

ctx.add_basemap(ax,crs='EPSG:4326', source=ctx.providers.Stamen.TonerLite, zoom=10)

plt.show()
```
