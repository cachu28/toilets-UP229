# Income in Los Angeles County

This notebook explores median income levels by census tract for Los Angeles County. We use American Community Survey 2019 5 year estimates and see the median income disproportion throughout the County. Finally, we compare this data to the locations of publicly available toilets throughout the County with an emphasis on the City of Los Angeles. 

```python
#import libraries
import pandas as pd
import geopandas as gpd
import contextily as ctx
import matplotlib.pyplot as plt
import json
import requests
```

```python
# call median income census data using api
# I will be using B06011_001E to call estimate of median household income for the past 12 months for 2019

r = requests.get('https://api.census.gov/data/2019/acs/acs5?get=NAME,B06011_001E&for=tract:*&in=state:06%20county:037')
censusdata = r.json()
census_income = pd.DataFrame(censusdata[1:], columns=censusdata[0])
census_income.head()
```

```python
#clean up data, drop unnecessary columns
census_income.drop (columns=['NAME', 'state', 'county'], inplace=True)
```

```python
#rename to something more understandable
census_income.rename(columns={'B06011_001E': 'median_income'}, inplace=True)
```

```python
#convert total_pop from object to integer
census_income = census_income.astype({'median_income':'float'})
```

```python
#set index in preparation for the join
census_income.set_index('tract', inplace=True)
```

```python
census_income.head()
```

```python
#next I want to check if there are any census tracts with NAN or anything less than 0 (so negative entries)
census_income[census_income<0]
```

```python
# since I have some that are NaN or negatice, I'm going to drop them and only keep values greater than 0
census_income= census_income[census_income>0]
census_income['median_income'] = census_income['median_income'].fillna(0)
census_income
```

# Census Tract Geometry Data

```python
#bring in census tracts which I downloaded from LA County Open Data Portal
#set crs too

tractsGdf = gpd.read_file('Census Tracts 2010.geojson').to_crs(epsg='4326')
```

```python
# slice the geoid column so that it matches the 'tract' column from our census_pop_df
# this is to prepare it for a join on the index

tractsGdf['short_geoid'] = tractsGdf['geoid10'].str.slice(5,11)
```

```python
tractsGdf.head()
```

```python
#set the index
tractsGdf.set_index('short_geoid', inplace=True)
```

# Join the Data

```python
# bring back census_race to join on index
incomeGdf = tractsGdf.join (census_income)
```

```python
incomeGdf.head()
```

```python
# prepping the data to join to all toliets, so I will reset the index
incomeGdf.reset_index(inplace=True)
```

```python
#change this column to an integer so i can join on integer later
incomeGdf = incomeGdf.astype({'short_geoid':'int'})
```

```python
#reset the index
incomeGdf.set_index('short_geoid', inplace=True)
```

```python
fig,ax = plt.subplots(figsize=(15,15))

# while lines to show block groups (sandwiched between the two layers above for visual clarity)
incomeGdf.boundary.plot(ax=ax,color='black',zorder=15,lw=0.5)
```

```python
#now I want to map the median income
incomeGdf.plot(figsize=(20,20),
         column='median_income',
         legend=True,
         scheme='equal_interval')
```

# Bring in Public Toilets Data

```python
#bring in all public toilets data
toilets = gpd.read_file('alltoilets.geojson').to_crs(epsg='4326')
```

```python
toilets.head()
```

```python
#LA county wide map

f, ax = plt.subplots(figsize=(20,20))

toilets.plot('type', ax=ax, marker='o', markersize=20, legend=True)

incomeGdf.plot('median_income', 
            ax=ax, 
            cmap='RdPu', 
            legend=True, 
            legend_kwds={'orientation': 'vertical'}, 
            alpha=0.3, 
            label='Count')
ax.set_xlim([-118.721783, -118.08]) 
ax.set_ylim([33.68, 34.4])
ax.set_title('Median Income with Santitation Stations in LA County', fontsize=25)

ctx.add_basemap(ax,crs='EPSG:4326', source=ctx.providers.Stamen.TonerLite)

plt.show()
```
