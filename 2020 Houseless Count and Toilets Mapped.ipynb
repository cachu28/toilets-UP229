## 2020 Houseless Count and Toilets Mapped

```python
import pandas as pd
import geopandas as gpd
import contextily as ctx
import matplotlib.pyplot as plt
```

```python
toilets = gpd.read_file('MiscData/toilets.geojson').to_crs(epsg=3857)
```

```python
toilets.sample(10)
```

```python
counts = gpd.read_file('MiscData/2020countbytract.geojson')
```

```python
counts.head()
```

```python
# Plotting both toilets and counts on the same map

f, ax = plt.subplots(figsize=(20,20))

toilets.plot('type', ax=ax, marker='o', markersize=20, legend=True)

counts.plot('Count', ax=ax, cmap='plasma', legend=True, legend_kwds={'orientation': 'horizontal'}, alpha=0.3, label='Count')

ax.set_xlim([-118.721783, -118.08]) 
ax.set_ylim([33.68, 34.4])

ctx.add_basemap(ax,crs='EPSG:4326', source=ctx.providers.Stamen.TonerLite)

plt.show()
```

```python
f, ax = plt.subplots()

counts.plot('Count', ax=ax, cmap='plasma', legend=True, legend_kwds={'orientation': 'horizontal'}, alpha=0.3, label='counts')

ax.set_xlim([-118.721783, -118.08]) 
ax.set_ylim([33.68, 34.4])
ax.legend(['counts'])

ctx.add_basemap(ax,crs='EPSG:4326', source=ctx.providers.Stamen.TonerLite)

plt.show()
```
