# Scraping the LACAN Google Map Layer

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.google.com/maps/d/u/0/viewer?mid=1ak8NcsKDwokcV0T9j0eQALz7rIHNWzwq&hl=en&ll=34.06844279585848%2C-118.27066246237175&z=10'
r = requests.get(url)

soup = BeautifulSoup(r.content)
print(soup.prettify())
```
